```python
from sklearn.preprocessing import RobustScaler  # Best for financial data with outliers
import joblib
import pandas as pd
```

```python
# Choose numeric features to scale (exclude target and identifiers)
numeric_features = [
    'amt',               # Transaction amount
    'city_pop',          # City population
    'distance_from_home',# Distance between home and merchant
    'time_since_last_txn'# Time since last transaction (seconds)
]
```

```python
def scale_train_data(train_df):
    """Safe feature scaling for training data with automatic feature validation"""
    # Make copy of dataframe
    df = train_df.copy()

    # 1. Identify numeric features
    numeric_features = ['amt', 'city_pop']
    optional_features = ['distance_from_home', 'time_since_last_txn']

    # Add optional features if they exist
    for feat in optional_features:
        if feat in df.columns:
            numeric_features.append(feat)

    # 2. Create time feature if needed
    if 'time_since_last_txn' not in df.columns and 'trans_date_trans_time' in df.columns:
        df = df.sort_values(['cc_num', 'trans_date_trans_time'])
        df['time_since_last_txn'] = df.groupby('cc_num')['trans_date_trans_time'].diff().dt.total_seconds().fillna(0)
        numeric_features.append('time_since_last_txn')

    # 3. Verify features exist
    missing = [col for col in numeric_features if col not in df.columns]
    if missing:
        raise KeyError(f"Cannot scale - missing features: {missing}")

    # 4. Apply Robust Scaling
    scaler = RobustScaler()
    df[numeric_features] = scaler.fit_transform(df[numeric_features])

    # 5. Save scaler
    joblib.dump(scaler, 'scaler.pkl')

    print(f"✅ Successfully scaled {len(numeric_features)} features")
    return df
```

```python
# Save for use on test data
joblib.dump(scaler, 'scaler.pkl')
print("✅ Scaler fitted and saved to scaler.pkl")
```

```python
def print_scaling_statistics(df, numeric_features):
    """Safe printing of scaling statistics with feature validation"""
    # Check which features actually exist in the DataFrame
    available_features = [col for col in numeric_features if col in df.columns]
    missing_features = set(numeric_features) - set(available_features)

    if missing_features:
        print(f"⚠️ Warning: Missing features {missing_features} - showing available features only")

    if not available_features:
        print("❌ No valid features available for statistics")
        return

    print("\n🔄 Scaled Data Statistics:")
    print(df[available_features].describe().round(2))

    # RobustScaler statistics
    print("\n📊 Median/Quartiles (RobustScaler output):")
    print("Medians:\n", df[available_features].median().round(2))
    print("\nIQR (Q3-Q1):\n",
          (df[available_features].quantile(0.75) - df[available_features].quantile(0.25)).round(2))
```

```python
import matplotlib.pyplot as plt

def plot_feature_distributions(df, numeric_features, scaler=None):
    """Safe boxplot visualization before/after scaling with feature validation"""
    # Check which features actually exist
    available_features = [col for col in numeric_features if col in df.columns]
    missing_features = set(numeric_features) - set(available_features)

    if missing_features:
        print(f"⚠️ Note: Missing features {missing_features} - plotting available features only")

    if not available_features:
        print("❌ No valid features available for plotting")
        return

    # Create figure
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

    # Original data boxplot
    df[available_features].plot(kind='box', ax=ax1)
    ax1.set_title('Original Features (Before Scaling)')
    ax1.set_ylabel('Value Range')

    # Scaled data boxplot (if scaler provided)
    if scaler:
        try:
            scaled_data = pd.DataFrame(
                scaler.transform(df[available_features]),
                columns=available_features
            )
            scaled_data.plot(kind='box', ax=ax2)
            ax2.set_title('Scaled Features (After Robust Scaling)')
            ax2.set_ylabel('Scaled Range')
        except Exception as e:
            print(f"❌ Scaling visualization failed: {str(e)}")
            ax2.set_visible(False)
    else:
        ax2.set_visible(False)
        print("ℹ️ No scaler provided - only showing original features")

    plt.tight_layout()
    plt.show()

# Usage example:
numeric_features = ['amt', 'city_pop', 'distance_from_home', 'time_since_last_txn']

# Before scaling
print("\n📊 Before Scaling:")
plot_feature_distributions(df, numeric_features)

# After scaling (if you've already fit the scaler)
if 'scaler' in locals():
    print("\n📊 After Scaling:")
    plot_feature_distributions(df, numeric_features, scaler)
else:
    print("\nℹ️ Run scaling first to see before/after comparison")
```
