```python
# Binary encode gender
df['gender'] = df['gender'].map({'M': 0, 'F': 1})

# One-Hot encode category
df = pd.get_dummies(df, columns=['category'], drop_first=True)
```

```python
# Target encoding for merchant/city/job
for col in ['merchant', 'city', 'job']:
    fraud_rate = df.groupby(col)['is_fraud'].mean()
    df[f'{col}_fraud_rate'] = df[col].map(fraud_rate)

# Hashing alternative for merchant
df['merchant_hash'] = df['merchant'].apply(lambda x: hash(x) % 1000)
```

```python
# Drop original categorical columns
df = df.drop(['merchant', 'city', 'job'], axis=1)

# Verify
print(df.filter(regex='_fraud_rate|_hash|category_|gender').head())
print("\nMissing values:", df.isnull().sum().sum())
```

```python
# Extract datetime features
df['hour'] = pd.to_datetime(df['trans_date_trans_time']).dt.hour
df['day_of_week'] = pd.to_datetime(df['trans_date_trans_time']).dt.dayofweek
```

```python
# Calculate transaction distance
df['distance_from_home'] = np.sqrt(
    (df['merch_lat'] - df['lat'])**2 +
    (df['merch_long'] - df['long'])**2
)
```
